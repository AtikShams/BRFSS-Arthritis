```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

# ==========================================
# 1. LOAD THE MAPPED DATASET
# ==========================================
data_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/2-BRFSS 2021 Feature maps.csv'
save_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'

print("Loading mapped dataset...")
df = pd.read_csv(data_path)
print(f"Dataset loaded successfully. Shape: {df.shape}")

# ==========================================
# 2. CALCULATE CORRELATION MATRIX
# ==========================================
print("Calculating correlation matrix...")
corr_matrix = df.corr(method='spearman')

# Create a mask to hide the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# ==========================================
# 3. PLOT AND SAVE THE HEATMAP (WITH VALUES)
# ==========================================
print("Generating heatmap with values...")

plt.figure(figsize=(24, 20))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# annot=True turns on the numbers
# fmt=".2f" rounds them to 2 decimal places
# annot_kws controls the font size and weight of the numbers inside the boxes
sns.heatmap(corr_matrix,
            mask=mask,
            cmap=cmap,
            vmax=1.0,
            vmin=-1.0,
            center=0,
            square=True,
            linewidths=.5,
            cbar_kws={"shrink": .75, "label": "Spearman Correlation Coefficient"},
            annot=True,
            fmt=".2f",
            annot_kws={"size": 9, "weight": "bold"})

# Formatting fonts to be bold and larger for publication
plt.title('Spearman Correlation Matrix of BRFSS 2021 Features',
          fontsize=22, fontweight='bold', pad=20)
plt.xticks(fontsize=12, fontweight='bold', rotation=45, ha='right')
plt.yticks(fontsize=12, fontweight='bold')

plt.tight_layout()

# Save the high-resolution image with a new name so it doesn't overwrite your blank one
save_path = os.path.join(save_dir, 'BRFSS_2021_Correlation_Heatmap_With_Values.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')

print(f"Correlation heatmap saved successfully to: {save_path}")

plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Load your perfectly mapped dataset
file_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/2-BRFSS 2021 Feature maps.csv'
df_mapped = pd.read_csv(file_path)

# 2. Drop rows where the target 'Arthritis' is missing
# (We cannot train or test on unknown targets)
df_clean = df_mapped.dropna(subset=['Arthritis'])
print(f"Rows before dropping NaN targets: {df_mapped.shape[0]}")
print(f"Rows after dropping NaN targets: {df_clean.shape[0]}\n")

# 3. Separate features (X) and target (y)
X = df_clean.drop(columns=['Arthritis'])
y = df_clean['Arthritis']

# 4. Perform the 80/20 Stratified Split
# stratify=y ensures both sets have the exact same percentage of Arthritis cases
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

print("=== SPLIT SUCCESSFUL ===")
print(f"Train Set Shape: {X_train.shape[0]} rows")
print(f"Test Set Shape: {X_test.shape[0]} rows\n")

# 5. Recombine X and y to save as full, intact datasets
train_df = X_train.copy()
train_df['Arthritis'] = y_train

test_df = X_test.copy()
test_df['Arthritis'] = y_test

# 6. Save the locked-in datasets to Google Drive
train_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/BRFSS_2021_Train.csv'
test_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/BRFSS_2021_Test.csv'

train_df.to_csv(train_path, index=False)
test_df.to_csv(test_path, index=False)

print(f"Train set securely saved to: {train_path}")
print(f"Test set securely saved to: {test_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp, wasserstein_distance
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings('ignore')

# ==========================================
# 1. LOAD THE LOCKED-IN SPLITS
# ==========================================
train_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/BRFSS_2021_Train.csv'
test_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/BRFSS_2021_Test.csv'

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

X_train = df_train.drop(columns=['Arthritis'])
y_train = df_train['Arthritis']
X_test = df_test.drop(columns=['Arthritis'])
y_test = df_test['Arthritis']

# ==========================================
# 2. STRICT ONE-HOT ENCODING (PREVENT LEAKAGE)
# ==========================================
columns_to_encode = [
    'Parents Separation', 'Race', 'Employment statues',
    'Marital status', 'Housing Status', 'Cholesterol Awareness'
]

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# FIT and TRANSFORM strictly on TRAIN
encoded_train = pd.DataFrame(encoder.fit_transform(X_train[columns_to_encode]))
encoded_train.columns = encoder.get_feature_names_out(columns_to_encode)
encoded_train.index = X_train.index

# TRANSFORM ONLY on TEST (Zero leakage)
encoded_test = pd.DataFrame(encoder.transform(X_test[columns_to_encode]))
encoded_test.columns = encoder.get_feature_names_out(columns_to_encode)
encoded_test.index = X_test.index

# Drop original nominal columns and safely concatenate encoded ones
X_train_enc = pd.concat([X_train.drop(columns=columns_to_encode), encoded_train], axis=1)
X_test_enc = pd.concat([X_test.drop(columns=columns_to_encode), encoded_test], axis=1)

print("=== ENCODING SUCCESSFUL ===")
print(f"Original features: {X_train.shape[1]}")
print(f"Features after One-Hot Encoding: {X_train_enc.shape[1]}\n")

# ==========================================
# 3. STATISTICAL EVALUATION FUNCTION
# ==========================================
def evaluate_imputation(original_df, imputed_df, feature_name, method_name):
    """Calculates KS-Test, Wasserstein Distance, and plots distributions."""

    # Drop NaNs from the original data to get the true baseline distribution
    original_data = original_df[feature_name].dropna()
    imputed_data = imputed_df[feature_name]

    # Calculate Statistical Metrics
    ks_stat, ks_pval = ks_2samp(original_data, imputed_data)
    wass_dist = wasserstein_distance(original_data, imputed_data)

    print(f"--- {method_name} | Feature: {feature_name} ---")
    print(f"Wasserstein Distance: {wass_dist:.4f} (Closer to 0 is better)")
    print(f"K-S Statistic: {ks_stat:.4f} (P-value: {ks_pval:.4e})\n")

    # Generate Density Plots for Visual Proof
    plt.figure(figsize=(8, 5))
    sns.kdeplot(original_data, label='Original (No NaNs)', color='blue', linewidth=2)
    sns.kdeplot(imputed_data, label=f'Imputed ({method_name})', color='red', linestyle="--", linewidth=2)
    plt.title(f'Distribution Comparison: {feature_name}\nMethod: {method_name} (W-Dist: {wass_dist:.4f})')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.tight_layout()
    plt.show()

print("Setup complete. Ready for imputation techniques.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp, wasserstein_distance
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense
import os
import warnings
warnings.filterwarnings('ignore')

# Define the path for the performance log
csv_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/imputation performance.csv'

# Define the directory to save the images and datasets
output_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/GAN filling method/'
os.makedirs(output_dir, exist_ok=True) # Ensures the folder exists

# Define the 3 features to evaluate (Must include 1 ACEs feature)
evaluation_features = ['Depressed Acompany', 'Income Level', 'BMI']

def evaluate_and_log_imputation(original_df, imputed_df, features_list, method_name, csv_file_path, img_save_dir):
    """Calculates metrics, plots distributions, saves images, and logs results."""

    results = []

    for feature_name in features_list:
        # Drop NaNs from original to get the true baseline
        original_data = original_df[feature_name].dropna()
        imputed_data = imputed_df[feature_name]

        # Calculate Metrics
        ks_stat, ks_pval = ks_2samp(original_data, imputed_data)
        wass_dist = wasserstein_distance(original_data, imputed_data)

        print(f"--- {method_name} | Feature: {feature_name} ---")
        print(f"Wasserstein Distance: {wass_dist:.4f} (Closer to 0 is better)")
        print(f"K-S Statistic: {ks_stat:.4f} (P-value: {ks_pval:.4e})\n")

        # Store for CSV
        results.append({
            'Imputation Technique': method_name,
            'Feature Evaluated': feature_name,
            'Wasserstein Distance': wass_dist,
            'K-S Statistic': ks_stat,
            'P-value': ks_pval
        })

        # Plotting
        plt.figure(figsize=(10, 6))
        sns.kdeplot(original_data, label='Original (No NaNs)', color='blue', linewidth=2.5)
        sns.kdeplot(imputed_data, label=f'Imputed ({method_name})', color='red', linestyle="--", linewidth=2.5)

        # Formatting fonts to be bold and larger
        plt.title(f'Distribution Comparison: {feature_name}\nMethod: {method_name} (W-Dist: {wass_dist:.4f})',
                  fontsize=16, fontweight='bold', pad=15)
        plt.xlabel('Value', fontsize=14, fontweight='bold', labelpad=10)
        plt.ylabel('Density', fontsize=14, fontweight='bold', labelpad=10)

        # Formatting legend
        plt.legend(prop={'size': 12, 'weight': 'bold'})

        plt.tight_layout()

        # Save the image
        safe_feature_name = feature_name.replace(" ", "_").replace("/", "-")
        img_filename = f"{method_name}_{safe_feature_name}_Distribution.png"
        img_save_path = os.path.join(img_save_dir, img_filename)
        plt.savefig(img_save_path, dpi=300, bbox_inches='tight') # High resolution save
        print(f"Saved plot to: {img_save_path}")

        plt.show()

    # Log to CSV
    result_df = pd.DataFrame(results)

    # Append to CSV (write header if file doesn't exist)
    if not os.path.isfile(csv_file_path):
        result_df.to_csv(csv_file_path, index=False)
    else:
        result_df.to_csv(csv_file_path, mode='a', header=False, index=False)
    print(f"\nMetrics for all {len(features_list)} features successfully logged to {csv_file_path}\n")


# ==========================================
# TECHNIQUE 8: GENERATIVE ADVERSARIAL NETWORK (GAN)
# ==========================================
print("Executing Technique 8: GAN Imputation...\n")

# 1. Temporarily fill NaNs with mean so the Network has a baseline to improve upon
temp_imputer = SimpleImputer(strategy='mean')
X_train_temp = temp_imputer.fit_transform(X_train_enc)
X_test_temp = temp_imputer.transform(X_test_enc)

# 2. Scale data to [0, 1] for neural network stability
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_temp)
X_test_scaled = scaler.transform(X_test_temp)

input_dim = X_train_scaled.shape[1]

# 3. Build the Generator
generator = Sequential([
    Input(shape=(input_dim,)),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(input_dim, activation='sigmoid') # Sigmoid bounds output between 0 and 1 to match MinMaxScaler
], name="Generator")

# 4. Build the Discriminator
discriminator = Sequential([
    Input(shape=(input_dim,)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid') # Outputs probability of being Real (1) or Fake (0)
], name="Discriminator")

discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 5. Build the Combined GAN Model
discriminator.trainable = False # Lock discriminator weights during generator training
gan_input = Input(shape=(input_dim,))
gan_output = discriminator(generator(gan_input))
gan = Model(gan_input, gan_output)
gan.compile(optimizer='adam', loss='binary_crossentropy')

# 6. Train the GAN (Custom Training Loop)
epochs = 15
batch_size = 256
half_batch = batch_size // 2

print(f"Training GAN for {epochs} epochs...")
for epoch in range(epochs):
    # Select a random half-batch of real data
    idx = np.random.randint(0, X_train_scaled.shape[0], half_batch)
    real_data = X_train_scaled[idx]

    # Generate a half-batch of fake data
    noise_idx = np.random.randint(0, X_train_scaled.shape[0], half_batch)
    noise_data = X_train_scaled[noise_idx] # Using temp-filled data as seed
    fake_data = generator.predict(noise_data, verbose=0)

    # Train the Discriminator
    d_loss_real = discriminator.train_on_batch(real_data, np.ones((half_batch, 1)))
    d_loss_fake = discriminator.train_on_batch(fake_data, np.zeros((half_batch, 1)))
    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

    # Train the Generator (via the combined model)
    # The generator wants the discriminator to think its fake data is real (ones)
    noise_for_g = X_train_scaled[np.random.randint(0, X_train_scaled.shape[0], batch_size)]
    g_loss = gan.train_on_batch(noise_for_g, np.ones((batch_size, 1)))

    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1}/{epochs} | D Loss: {d_loss[0]:.4f} | G Loss: {g_loss:.4f}")

# 7. Predict (Generate the final synthetic data)
X_train_pred_scaled = generator.predict(X_train_scaled)
X_test_pred_scaled = generator.predict(X_test_scaled)

# 8. Inverse transform back to original numerical scale
X_train_pred_np = scaler.inverse_transform(X_train_pred_scaled)
X_test_pred_np = scaler.inverse_transform(X_test_pred_scaled)

# Convert to DataFrames
X_train_pred = pd.DataFrame(X_train_pred_np, columns=X_train_enc.columns, index=X_train_enc.index)
X_test_pred = pd.DataFrame(X_test_pred_np, columns=X_test_enc.columns, index=X_test_enc.index)

# 9. Targeted Replacement: Only overwrite the actual missing values
X_train_gan = X_train_enc.copy()
X_test_gan = X_test_enc.copy()

mask_train = X_train_enc.isnull()
mask_test = X_test_enc.isnull()

X_train_gan[mask_train] = X_train_pred[mask_train]
X_test_gan[mask_test] = X_test_pred[mask_test]

# Evaluate and log
evaluate_and_log_imputation(X_train_enc, X_train_gan, evaluation_features, 'GAN', csv_path, output_dir)

# Save Imputed Datasets
train_save_path = os.path.join(output_dir, 'Train_Imputed_GAN.csv')
test_save_path = os.path.join(output_dir, 'Test_Imputed_GAN.csv')

X_train_gan.to_csv(train_save_path, index=False)
X_test_gan.to_csv(test_save_path, index=False)

print(f"Saved imputed train set to: {train_save_path}")
print(f"Saved imputed test set to: {test_save_path}")
```

# **Model training (GB, RF, LGBM, LR, XGBOOST ) for xgboost+ROS**

```python
import pandas as pd
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import RandomizedSearchCV, ShuffleSplit, cross_validate
from sklearn.metrics import classification_report, balanced_accuracy_score, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# ==========================================
# 1. SETUP PATHS AND LOAD DATA
# ==========================================
print("Loading datasets...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'
train_path = os.path.join(base_dir, 'XGBoost filling method/Resampled Datasets/3-BRFSS 2021 - Fill Xgboost ROS (processed).csv')
test_features_path = os.path.join(base_dir, 'XGBoost filling method/Test_Imputed_XGBoostFilling (processed).csv')
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')

# Create directories for saving outputs
cm_dir = os.path.join(base_dir, 'XGBoost filling method/Confusion Matrices/')
os.makedirs(cm_dir, exist_ok=True)
excel_output_path = os.path.join(base_dir, 'Model_Performance_XGBoost_ROS.xlsx')

# Load Train
df_train = pd.read_csv(train_path)
X_train = df_train.drop('Arthritis', axis=1)
y_train = df_train['Arthritis']

# Load Test Features and Target
X_test = pd.read_csv(test_features_path)
df_test_orig = pd.read_csv(original_test_path)
y_test = df_test_orig['Arthritis']

# Clean up any potential decimal artifacts in X_test
X_test = X_test.round().astype(int)

# String identifiers
data_name = "XGBoost Imputation"
sampling_name = "ROS"

# Initialize Performance Dictionaries
performance = {
    'Fill method': [], 'Model': [], 'Sampling': [], 'Model Sampling': [],
    'Class 0 #': [], 'Class 1 #': [],
    'Precision c1': [], 'Recall c1': [], 'F1 Score c1': [], 'Support c1': [],
    'Precision c0': [], 'Recall c0': [], 'F1 Score c0': [], 'Support c0': [],
    'Balanced Accuracy': [], 'AUROC': []
}

cv_performance = {
    'Model': [], 'Best Parameters': [],
    'Fold 1 (AUROC)': [], 'Fold 2 (AUROC)': [], 'Fold 3 (AUROC)': [],
    'Fold 4 (AUROC)': [], 'Fold 5 (AUROC)': [],
    'Mean CV AUROC': [], 'Std CV AUROC': []
}

# ==========================================
# 2. DEFINE MODELS AND HYPERPARAMETER GRIDS
# ==========================================
models = {
    "Logistic Regression": {
        "model": LogisticRegression(random_state=42, max_iter=1000, n_jobs=-1),
        "params": {'C': [0.01, 0.1, 1, 10], 'penalty': ['l2']}
    },
    "Random Forest": {
        "model": RandomForestClassifier(random_state=42, n_jobs=-1),
        "params": {'n_estimators': [100, 200], 'max_depth': [10, 20, None]}
    },
    "Gradient Boosting": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 5]}
    },
    "XGBoost": {
        "model": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', n_jobs=-1),
        "params": {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 5, 7]}
    },
    "LightGBM": {
        "model": LGBMClassifier(random_state=42, n_jobs=-1),
        "params": {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [-1, 10, 20]}
    }
}

# ==========================================
# 3. TRAIN, TUNE, AND EVALUATE (WITH SAFE-SAVE)
# ==========================================
fast_tune_split = ShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for model_name, mp in models.items():
    print(f"\n[{model_name}] Step 1: Fast Hyperparameter Tuning...")
    start_time = time.time()

    clf = RandomizedSearchCV(mp['model'], mp['params'], n_iter=5, cv=fast_tune_split,
                             scoring='roc_auc', random_state=42, n_jobs=-1, refit=True)

    clf.fit(X_train, y_train)
    best_model = clf.best_estimator_

    print(f"[{model_name}] Tuning done in {(time.time() - start_time)/60:.2f} minutes.")
    print(f"[{model_name}] Best Parameters: {clf.best_params_}")

    # --- Step 2: 5-Fold Cross Validation on the Best Model ---
    print(f"[{model_name}] Step 2: Running 5-Fold CV on the best model...")
    cv_time = time.time()
    cv_res = cross_validate(best_model, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1)

    folds = cv_res['test_score']
    mean_cv = folds.mean()
    std_cv = folds.std()
    print(f"[{model_name}] CV done in {(time.time() - cv_time)/60:.2f} minutes.")

    cv_performance['Model'].append(model_name)
    cv_performance['Best Parameters'].append(str(clf.best_params_))
    cv_performance['Fold 1 (AUROC)'].append(folds[0])
    cv_performance['Fold 2 (AUROC)'].append(folds[1])
    cv_performance['Fold 3 (AUROC)'].append(folds[2])
    cv_performance['Fold 4 (AUROC)'].append(folds[3])
    cv_performance['Fold 5 (AUROC)'].append(folds[4])
    cv_performance['Mean CV AUROC'].append(mean_cv)
    cv_performance['Std CV AUROC'].append(std_cv)

    # --- Step 3: Predict on Test Set ---
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1]

    # --- Confusion Matrix Plotting & Saving ---
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 14, "weight": "bold"})
    plt.title(f'{model_name} Confusion Matrix\n({sampling_name})', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')
    plt.tight_layout()

    cm_filename = f"CM_{model_name.replace(' ', '_')}_{sampling_name}.png"
    plt.savefig(os.path.join(cm_dir, cm_filename), dpi=300)
    plt.close()

    # --- Calculate Metrics ---
    report = classification_report(y_test, y_pred, output_dict=True)
    balanced_acc = balanced_accuracy_score(y_test, y_pred)
    AUROC = roc_auc_score(y_test, y_pred_proba)

    train_counts = y_train.value_counts()

    precision_c0 = report['0.0']['precision']
    recall_c0 = report['0.0']['recall']
    f1_score_c0 = report['0.0']['f1-score']
    support_c0 = report['0.0']['support']

    precision_c1 = report['1.0']['precision']
    recall_c1 = report['1.0']['recall']
    f1_score_c1 = report['1.0']['f1-score']
    support_c1 = report['1.0']['support']

    train_class_0_size = train_counts.get(0.0, 0)
    train_class_1_size = train_counts.get(1.0, 0)

    # --- Print to Console ---
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
    print("Fill method:", data_name)
    print("Model:", model_name)
    print("Sampling:", sampling_name)
    print("Model Sampling:", model_name + ' ' + sampling_name)
    print("Class 0 #:", train_class_0_size)
    print("Class 1 #:", train_class_1_size)
    print("Precision c1:", precision_c1)
    print("Recall c1:", recall_c1)
    print("F1 Score c1:", f1_score_c1)
    print("Support c1:", support_c1)
    print("Precision c0:", precision_c0)
    print("Recall c0:", recall_c0)
    print("F1 Score c0:", f1_score_c0)
    print("Support c0:", support_c0)
    print("Balanced Accuracy:", balanced_acc)
    print("AUROC:", AUROC)
    print(f"Mean Train CV AUROC: {mean_cv:.4f} (+/- {std_cv:.4f})")
    print(f"Saved Confusion Matrix to Drive.")

    performance['Fill method'].append(data_name)
    performance['Model'].append(model_name)
    performance['Sampling'].append(sampling_name)
    performance['Model Sampling'].append(model_name + ' ' + sampling_name)
    performance['Class 0 #'].append(train_class_0_size)
    performance['Class 1 #'].append(train_class_1_size)
    performance['Precision c1'].append(precision_c1)
    performance['Recall c1'].append(recall_c1)
    performance['F1 Score c1'].append(f1_score_c1)
    performance['Support c1'].append(support_c1)
    performance['Precision c0'].append(precision_c0)
    performance['Recall c0'].append(recall_c0)
    performance['F1 Score c0'].append(f1_score_c0)
    performance['Support c0'].append(support_c0)
    performance['Balanced Accuracy'].append(balanced_acc)
    performance['AUROC'].append(AUROC)

    # ==========================================
    # 4. SAVE PROGRESS TO EXCEL (INSIDE THE LOOP)
    # ==========================================
    perf_df = pd.DataFrame(performance)
    cv_df = pd.DataFrame(cv_performance)

    with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
        perf_df.to_excel(writer, sheet_name='Test Metrics', index=False)
        cv_df.to_excel(writer, sheet_name='CV Scores', index=False)

    print(f"[{model_name}] Progress fully saved to Excel.\n")

print("All models successfully completed!")
```

```python
import pandas as pd
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import RandomizedSearchCV, ShuffleSplit, cross_validate
from sklearn.metrics import classification_report, balanced_accuracy_score, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# ==========================================
# 1. SETUP PATHS AND LOAD DATA
# ==========================================
print("Loading datasets...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'
train_path = os.path.join(base_dir, 'XGBoost filling method/Resampled Datasets/3-BRFSS 2021 - Fill Xgboost ROS (processed).csv')
test_features_path = os.path.join(base_dir, 'XGBoost filling method/Test_Imputed_XGBoostFilling (processed).csv')
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')

# Create directories for saving outputs
cm_dir = os.path.join(base_dir, 'XGBoost filling method/Confusion Matrices/')
os.makedirs(cm_dir, exist_ok=True)
excel_output_path = os.path.join(base_dir, 'Model_Performance_XGBoost_ROS.xlsx')

# Load Train
df_train = pd.read_csv(train_path)
X_train = df_train.drop('Arthritis', axis=1)
y_train = df_train['Arthritis']

# Load Test Features and Target
X_test = pd.read_csv(test_features_path)
df_test_orig = pd.read_csv(original_test_path)
y_test = df_test_orig['Arthritis']

# Clean up any potential decimal artifacts in X_test
X_test = X_test.round().astype(int)

# String identifiers
data_name = "XGBoost Imputation"
sampling_name = "ROS"

# Initialize Performance Dictionaries
performance = {
    'Fill method': [], 'Model': [], 'Sampling': [], 'Model Sampling': [],
    'Class 0 #': [], 'Class 1 #': [],
    'Precision c1': [], 'Recall c1': [], 'F1 Score c1': [], 'Support c1': [],
    'Precision c0': [], 'Recall c0': [], 'F1 Score c0': [], 'Support c0': [],
    'Balanced Accuracy': [], 'AUROC': []
}

cv_performance = {
    'Model': [], 'Best Parameters': [],
    'Fold 1 (AUROC)': [], 'Fold 2 (AUROC)': [], 'Fold 3 (AUROC)': [],
    'Fold 4 (AUROC)': [], 'Fold 5 (AUROC)': [],
    'Mean CV AUROC': [], 'Std CV AUROC': []
}

# ==========================================
# 2. DEFINE MODELS AND HYPERPARAMETER GRIDS
# ==========================================
models = {

    "Gradient Boosting": {
        "model": GradientBoostingClassifier(random_state=42),
        "params": {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 5]}
    },
    "XGBoost": {
        "model": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', n_jobs=-1),
        "params": {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [3, 5, 7]}
    },
    "LightGBM": {
        "model": LGBMClassifier(random_state=42, n_jobs=-1),
        "params": {'n_estimators': [100, 200], 'learning_rate': [0.01, 0.1], 'max_depth': [-1, 10, 20]}
    },
     "Random Forest": {
        "model": RandomForestClassifier(random_state=42, n_jobs=-1),
        "params": {'n_estimators': [100, 200], 'max_depth': [10, 20, None]}
    }
}

# ==========================================
# 3. TRAIN, TUNE, AND EVALUATE (WITH SAFE-SAVE)
# ==========================================
fast_tune_split = ShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for model_name, mp in models.items():
    print(f"\n[{model_name}] Step 1: Fast Hyperparameter Tuning...")
    start_time = time.time()

    clf = RandomizedSearchCV(mp['model'], mp['params'], n_iter=5, cv=fast_tune_split,
                             scoring='roc_auc', random_state=42, n_jobs=-1, refit=True)

    clf.fit(X_train, y_train)
    best_model = clf.best_estimator_

    print(f"[{model_name}] Tuning done in {(time.time() - start_time)/60:.2f} minutes.")
    print(f"[{model_name}] Best Parameters: {clf.best_params_}")

    # --- Step 2: 5-Fold Cross Validation on the Best Model ---
    print(f"[{model_name}] Step 2: Running 5-Fold CV on the best model...")
    cv_time = time.time()
    cv_res = cross_validate(best_model, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1)

    folds = cv_res['test_score']
    mean_cv = folds.mean()
    std_cv = folds.std()
    print(f"[{model_name}] CV done in {(time.time() - cv_time)/60:.2f} minutes.")

    cv_performance['Model'].append(model_name)
    cv_performance['Best Parameters'].append(str(clf.best_params_))
    cv_performance['Fold 1 (AUROC)'].append(folds[0])
    cv_performance['Fold 2 (AUROC)'].append(folds[1])
    cv_performance['Fold 3 (AUROC)'].append(folds[2])
    cv_performance['Fold 4 (AUROC)'].append(folds[3])
    cv_performance['Fold 5 (AUROC)'].append(folds[4])
    cv_performance['Mean CV AUROC'].append(mean_cv)
    cv_performance['Std CV AUROC'].append(std_cv)

    # --- Step 3: Predict on Test Set ---
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1]

    # --- Confusion Matrix Plotting & Saving ---
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 14, "weight": "bold"})
    plt.title(f'{model_name} Confusion Matrix\n({sampling_name})', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')
    plt.tight_layout()

    cm_filename = f"CM_{model_name.replace(' ', '_')}_{sampling_name}.png"
    plt.savefig(os.path.join(cm_dir, cm_filename), dpi=300)
    plt.close()

    # --- Calculate Metrics ---
    report = classification_report(y_test, y_pred, output_dict=True)
    balanced_acc = balanced_accuracy_score(y_test, y_pred)
    AUROC = roc_auc_score(y_test, y_pred_proba)

    train_counts = y_train.value_counts()

    precision_c0 = report['0.0']['precision']
    recall_c0 = report['0.0']['recall']
    f1_score_c0 = report['0.0']['f1-score']
    support_c0 = report['0.0']['support']

    precision_c1 = report['1.0']['precision']
    recall_c1 = report['1.0']['recall']
    f1_score_c1 = report['1.0']['f1-score']
    support_c1 = report['1.0']['support']

    train_class_0_size = train_counts.get(0.0, 0)
    train_class_1_size = train_counts.get(1.0, 0)

    # --- Print to Console ---
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
    print("Fill method:", data_name)
    print("Model:", model_name)
    print("Sampling:", sampling_name)
    print("Model Sampling:", model_name + ' ' + sampling_name)
    print("Class 0 #:", train_class_0_size)
    print("Class 1 #:", train_class_1_size)
    print("Precision c1:", precision_c1)
    print("Recall c1:", recall_c1)
    print("F1 Score c1:", f1_score_c1)
    print("Support c1:", support_c1)
    print("Precision c0:", precision_c0)
    print("Recall c0:", recall_c0)
    print("F1 Score c0:", f1_score_c0)
    print("Support c0:", support_c0)
    print("Balanced Accuracy:", balanced_acc)
    print("AUROC:", AUROC)
    print(f"Mean Train CV AUROC: {mean_cv:.4f} (+/- {std_cv:.4f})")
    print(f"Saved Confusion Matrix to Drive.")

    performance['Fill method'].append(data_name)
    performance['Model'].append(model_name)
    performance['Sampling'].append(sampling_name)
    performance['Model Sampling'].append(model_name + ' ' + sampling_name)
    performance['Class 0 #'].append(train_class_0_size)
    performance['Class 1 #'].append(train_class_1_size)
    performance['Precision c1'].append(precision_c1)
    performance['Recall c1'].append(recall_c1)
    performance['F1 Score c1'].append(f1_score_c1)
    performance['Support c1'].append(support_c1)
    performance['Precision c0'].append(precision_c0)
    performance['Recall c0'].append(recall_c0)
    performance['F1 Score c0'].append(f1_score_c0)
    performance['Support c0'].append(support_c0)
    performance['Balanced Accuracy'].append(balanced_acc)
    performance['AUROC'].append(AUROC)

    # ==========================================
    # 4. SAVE PROGRESS TO EXCEL (INSIDE THE LOOP)
    # ==========================================
    perf_df = pd.DataFrame(performance)
    cv_df = pd.DataFrame(cv_performance)

    with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
        perf_df.to_excel(writer, sheet_name='Test Metrics', index=False)
        cv_df.to_excel(writer, sheet_name='CV Scores', index=False)

    print(f"[{model_name}] Progress fully saved to Excel.\n")

print("All models successfully completed!")
```

# **GAN+ROS data processing**

```python
import pandas as pd
import numpy as np
import os

# ==========================================
# 1. SETUP PATHS
# ==========================================
train_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/GAN filling method/Resampled Datasets/Train_GAN_OverSampling.csv'
test_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/GAN filling method/Test_Imputed_GAN.csv'

train_save_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/GAN filling method/Resampled Datasets/3-BRFSS 2021 - Fill GAN ROS (processed).csv'
test_save_path = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/GAN filling method/Test_Imputed_GAN_Processed.csv'

# ==========================================
# 2. DEFINE CLIPPING LIMITS (From your logic)
# ==========================================
feature_limits = {
    'Physically hurt': (0, 2),
    'Depressed Acompany': (0, 1),
    'Verbally Insulted': (0, 2),
    'Touched Sexually': (0, 2),
    'Sexual harassment': (0, 2),
    'Drug Addicted F_member': (0, 1),
    'Age': (1, 13),
    'Gender': (0, 1),
    'Income Level': (1, 11),
    'Education level': (0, 5),
    'Smoking status': (0, 1),
    'Alcohol consumption': (0, 30),
    'Physical activity': (0, 1),
    'Regular Checkup': (0, 4),
    'Dietary Habits (Fruits)': (0, 3),
    'Dietary Habits (Vegetables)': (0, 3),
    'DEAF': (0, 1),
    'BMI': (1, 4),
    'BLIND': (0, 1),
    'Difficulty concentrating': (0, 1),
    'DIFFWALK': (0, 1),
    'Health Insurance': (0, 1),
    'Mental Health': (0, 2),
    'Physical Health': (0, 2),
    'Blood pressure': (0, 1),
    'Arthritis': (0, 1) # Target variable
}

def process_and_verify_dataset(filepath, savepath, dataset_name):
    print(f"\n{'='*50}")
    print(f"PROCESSING {dataset_name.upper()} DATASET")
    print(f"{'='*50}")

    # Load dataset
    df = pd.read_csv(filepath)
    print(f"Loaded {dataset_name} shape: {df.shape}")

    # Check for decimals before processing
    decimals_before = any([df[col].dtype == 'float64' for col in df.columns])
    print(f"Decimals present before processing? {'Yes' if decimals_before else 'No'}")

    # List to store the features we want to print value counts for
    features_to_print = ['Age', 'Income Level', 'BMI', 'Depressed Acompany']

    print("\n--- SAMPLE STATUS BEFORE PROCESSING ---")
    for f in features_to_print:
        if f in df.columns:
            print(f"\n{f}:")
            print(df[f].value_counts().sort_index().head(5)) # Printing top 5 to save space

    # ==========================================
    # CORE PROCESSING: ROUND -> INT -> CLIP
    # ==========================================
    # 1. Round to nearest whole number and convert completely to integer
    df_clean = df.round().astype(int)

    # 2. Apply lower and upper limits using pandas .clip()
    for feature, (lower, upper) in feature_limits.items():
        if feature in df_clean.columns:
            # .clip mathematically limits values to the defined minimum and maximum boundaries
            df_clean[feature] = df_clean[feature].clip(lower=lower, upper=upper)

    print("\n--- SAMPLE STATUS AFTER PROCESSING (BOUNDARIES ENFORCED) ---")
    for f in features_to_print:
        if f in df.columns:
            print(f"\n{f}:")
            print(df_clean[f].value_counts().sort_index())

    # Verification Step
    decimals_after = any([df_clean[col].dtype == 'float64' for col in df_clean.columns])
    out_of_bounds = False

    for feature, (lower, upper) in feature_limits.items():
        if feature in df_clean.columns:
            if df_clean[feature].min() < lower or df_clean[feature].max() > upper:
                out_of_bounds = True

    print("\n--- FINAL VERIFICATION ---")
    print(f"Are there ANY decimal values left? {'Yes' if decimals_after else 'No (All Integers)'}")
    print(f"Are ANY values out of bounds? {'Yes' if out_of_bounds else 'No (Perfectly Bounded)'}")

    # Save processed dataset
    df_clean.to_csv(savepath, index=False)
    print(f"\nSaved processed {dataset_name} set to: {savepath}")

    return df_clean

# Run the function on Train Set
train_processed = process_and_verify_dataset(train_path, train_save_path, "Train")

# Run the function on Test Set
test_processed = process_and_verify_dataset(test_path, test_save_path, "Test")

print("\nData processing successfully completed for both datasets!")
```

```python
import pandas as pd
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import cross_validate
from sklearn.metrics import classification_report, balanced_accuracy_score, roc_auc_score, confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# ==========================================
# 1. SETUP PATHS AND LOAD DATA
# ==========================================
print("Loading GAN datasets...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'

train_path = os.path.join(base_dir, 'GAN filling method/Resampled Datasets/3-BRFSS 2021 - Fill GAN ROS (processed).csv')
test_features_path = os.path.join(base_dir, 'GAN filling method/Test_Imputed_GAN_Processed.csv')
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')

# Directories
cm_dir = os.path.join(base_dir, 'GAN filling method/Confusion Matrices/')
os.makedirs(cm_dir, exist_ok=True)
excel_output_path = os.path.join(base_dir, 'Performance_of_all_models.xlsx')

# Load Train & Test
df_train = pd.read_csv(train_path)
X_train = df_train.drop('Arthritis', axis=1)
y_train = df_train['Arthritis']

X_test = pd.read_csv(test_features_path)
df_test_orig = pd.read_csv(original_test_path)
y_test = df_test_orig['Arthritis']

data_name = "GAN Imputation"
sampling_name = "ROS"

# ==========================================
# 2. LOAD EXISTING EXCEL FILE FOR APPENDING
# ==========================================
# We load the historical data so we don't overwrite the XGBoost/LightGBM runs
if os.path.exists(excel_output_path):
    print("Found existing Performance Excel file. New results will be appended.")
    base_perf_df = pd.read_excel(excel_output_path, sheet_name='Test Metrics')
    base_cv_df = pd.read_excel(excel_output_path, sheet_name='CV Scores')
else:
    print("No existing Excel file found. A new one will be created.")
    base_perf_df = pd.DataFrame()
    base_cv_df = pd.DataFrame()

# Initialize dictionaries ONLY for the new models running now
performance = {
    'Fill method': [], 'Model': [], 'Sampling': [], 'Model Sampling': [],
    'Class 0 #': [], 'Class 1 #': [],
    'Precision c1': [], 'Recall c1': [], 'F1 Score c1': [], 'Support c1': [],
    'Precision c0': [], 'Recall c0': [], 'F1 Score c0': [], 'Support c0': [],
    'Overall Sensitivity': [], 'Overall Specificity': [],
    'Balanced Accuracy': [], 'AUROC': []
}

cv_performance = {
    'Model': [], 'Parameters': [],
    'Fold 1 (AUROC)': [], 'Fold 2 (AUROC)': [], 'Fold 3 (AUROC)': [],
    'Fold 4 (AUROC)': [], 'Fold 5 (AUROC)': [],
    'Mean CV AUROC': [], 'Std CV AUROC': []
}

# ==========================================
# 3. DEFINE NEW MODELS
# ==========================================
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42)
}

# ==========================================
# 4. TRAIN AND EVALUATE (WITH SAFE-SAVE APPENDING)
# ==========================================
for model_name, model in models.items():
    print(f"\n[{model_name}] Step 1: Running 5-Fold CV on Train Set...")
    cv_time = time.time()

    cv_res = cross_validate(model, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1)

    folds = cv_res['test_score']
    mean_cv = folds.mean()
    std_cv = folds.std()
    print(f"[{model_name}] CV done in {(time.time() - cv_time)/60:.2f} minutes.")

    cv_performance['Model'].append(model_name)
    cv_performance['Parameters'].append("Default")
    cv_performance['Fold 1 (AUROC)'].append(folds[0])
    cv_performance['Fold 2 (AUROC)'].append(folds[1])
    cv_performance['Fold 3 (AUROC)'].append(folds[2])
    cv_performance['Fold 4 (AUROC)'].append(folds[3])
    cv_performance['Fold 5 (AUROC)'].append(folds[4])
    cv_performance['Mean CV AUROC'].append(mean_cv)
    cv_performance['Std CV AUROC'].append(std_cv)

    # --- Step 2: Train Final Model and Predict on Test Set ---
    print(f"[{model_name}] Step 2: Training final model for Test Set prediction...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    # --- Confusion Matrix Plotting & Saving ---
    cm = confusion_matrix(y_test, y_pred)

    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 14, "weight": "bold"})
    plt.title(f'{model_name} Confusion Matrix\n({sampling_name})', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')
    plt.tight_layout()

    cm_filename = f"CM_{model_name.replace(' ', '_')}_{sampling_name}.png"
    plt.savefig(os.path.join(cm_dir, cm_filename), dpi=300)
    plt.close()

    # --- Calculate Metrics ---
    report = classification_report(y_test, y_pred, output_dict=True)
    balanced_acc = balanced_accuracy_score(y_test, y_pred)
    AUROC = roc_auc_score(y_test, y_pred_proba)

    train_counts = y_train.value_counts()

    precision_c0 = report['0.0']['precision']
    recall_c0 = report['0.0']['recall']
    f1_score_c0 = report['0.0']['f1-score']
    support_c0 = report['0.0']['support']

    precision_c1 = report['1.0']['precision']
    recall_c1 = report['1.0']['recall']
    f1_score_c1 = report['1.0']['f1-score']
    support_c1 = report['1.0']['support']

    train_class_0_size = train_counts.get(0.0, 0)
    train_class_1_size = train_counts.get(1.0, 0)

    # --- Print to Console ---
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
    print("Fill method:", data_name)
    print("Model:", model_name)
    print("Sampling:", sampling_name)
    print("Model Sampling:", model_name + ' ' + sampling_name)
    print("Class 0 #:", train_class_0_size)
    print("Class 1 #:", train_class_1_size)
    print("Precision c1:", precision_c1)
    print("Recall c1:", recall_c1)
    print("F1 Score c1:", f1_score_c1)
    print("Support c1:", support_c1)
    print("Precision c0:", precision_c0)
    print("Recall c0:", recall_c0)
    print("F1 Score c0:", f1_score_c0)
    print("Support c0:", support_c0)
    print("Overall Sensitivity:", sensitivity)
    print("Overall Specificity:", specificity)
    print("Balanced Accuracy:", balanced_acc)
    print("AUROC:", AUROC)
    print(f"Mean Train CV AUROC: {mean_cv:.4f} (+/- {std_cv:.4f})")
    print(f"Saved Confusion Matrix to Drive.")

    performance['Fill method'].append(data_name)
    performance['Model'].append(model_name)
    performance['Sampling'].append(sampling_name)
    performance['Model Sampling'].append(model_name + ' ' + sampling_name)
    performance['Class 0 #'].append(train_class_0_size)
    performance['Class 1 #'].append(train_class_1_size)
    performance['Precision c1'].append(precision_c1)
    performance['Recall c1'].append(recall_c1)
    performance['F1 Score c1'].append(f1_score_c1)
    performance['Support c1'].append(support_c1)
    performance['Precision c0'].append(precision_c0)
    performance['Recall c0'].append(recall_c0)
    performance['F1 Score c0'].append(f1_score_c0)
    performance['Support c0'].append(support_c0)
    performance['Overall Sensitivity'].append(sensitivity)
    performance['Overall Specificity'].append(specificity)
    performance['Balanced Accuracy'].append(balanced_acc)
    performance['AUROC'].append(AUROC)

    # ==========================================
    # 5. CONCATENATE & SAVE TO EXCEL (INSIDE LOOP)
    # ==========================================
    # Convert current run data to dataframes
    new_perf_df = pd.DataFrame(performance)
    new_cv_df = pd.DataFrame(cv_performance)

    # Combine historical base data with new data
    if not base_perf_df.empty:
        save_perf_df = pd.concat([base_perf_df, new_perf_df], ignore_index=True)
        save_cv_df = pd.concat([base_cv_df, new_cv_df], ignore_index=True)
    else:
        save_perf_df = new_perf_df
        save_cv_df = new_cv_df

    with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
        save_perf_df.to_excel(writer, sheet_name='Test Metrics', index=False)
        save_cv_df.to_excel(writer, sheet_name='CV Scores', index=False)

    print(f"[{model_name}] Progress fully appended and saved to Excel.\n")

print("All new models successfully completed and appended!")
```

# **Model training (ANN[1L], ANN[2L], Multihead Attention+ANN) for GAN+ROS**

```python
import pandas as pd
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report, balanced_accuracy_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler

import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Dense, Dropout, MultiHeadAttention, Reshape, Flatten

# ==========================================
# 1. SETUP PATHS AND LOAD DATA
# ==========================================
print("Loading GAN datasets for Deep Learning...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'

train_path = os.path.join(base_dir, 'GAN filling method/Resampled Datasets/3-BRFSS 2021 - Fill GAN ROS (processed).csv')
test_features_path = os.path.join(base_dir, 'GAN filling method/Test_Imputed_GAN_Processed.csv')
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')

cm_dir = os.path.join(base_dir, 'GAN filling method/Confusion Matrices/')
os.makedirs(cm_dir, exist_ok=True)
excel_output_path = os.path.join(base_dir, 'Performance_of_all_models.xlsx')

# Load Train & Test
df_train = pd.read_csv(train_path)
X_train = df_train.drop('Arthritis', axis=1).values # Convert to numpy array for Keras
y_train = df_train['Arthritis'].values

df_test = pd.read_csv(test_features_path)
X_test = df_test.values
df_test_orig = pd.read_csv(original_test_path)
y_test = df_test_orig['Arthritis'].values

data_name = "GAN Imputation"
sampling_name = "ROS"

# ==========================================
# 2. FEATURE SCALING (CRITICAL FOR NEURAL NETS)
# ==========================================
print("Scaling features for Neural Networks...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==========================================
# 3. LOAD EXISTING EXCEL FILE FOR APPENDING
# ==========================================
if os.path.exists(excel_output_path):
    print("Found existing Performance Excel file. New results will be appended.")
    base_perf_df = pd.read_excel(excel_output_path, sheet_name='Test Metrics')
    base_cv_df = pd.read_excel(excel_output_path, sheet_name='CV Scores')
else:
    print("No existing Excel file found. A new one will be created.")
    base_perf_df = pd.DataFrame()
    base_cv_df = pd.DataFrame()

# Initialize dictionaries
performance = {
    'Fill method': [], 'Model': [], 'Sampling': [], 'Model Sampling': [],
    'Class 0 #': [], 'Class 1 #': [],
    'Precision c1': [], 'Recall c1': [], 'F1 Score c1': [], 'Support c1': [],
    'Precision c0': [], 'Recall c0': [], 'F1 Score c0': [], 'Support c0': [],
    'Overall Sensitivity': [], 'Overall Specificity': [],
    'Balanced Accuracy': [], 'AUROC': []
}

cv_performance = {
    'Model': [], 'Parameters': [],
    'Fold 1 (AUROC)': [], 'Fold 2 (AUROC)': [], 'Fold 3 (AUROC)': [],
    'Fold 4 (AUROC)': [], 'Fold 5 (AUROC)': [],
    'Mean CV AUROC': [], 'Std CV AUROC': []
}

# ==========================================
# 4. DEFINE NEURAL NETWORK ARCHITECTURES
# ==========================================
input_dim = X_train_scaled.shape[1]

def build_ann_1_layer():
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

def build_ann_2_layer():
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

def build_attention_ann():
    inputs = Input(shape=(input_dim,))
    # Reshape for Attention Layer (Batch, Seq_Length, Features)
    x_reshaped = Reshape((input_dim, 1))(inputs)
    # Self-Attention Layer
    attention_out = MultiHeadAttention(num_heads=2, key_dim=2)(x_reshaped, x_reshaped)
    x = Flatten()(attention_out)
    # Feed into ANN 2 Layer
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

models = {
    "ANN 1 Layer": build_ann_1_layer,
    "ANN 2 Layer": build_ann_2_layer,
    "Attention + ANN 2 Layer": build_attention_ann
}

# Training Hyperparameters
EPOCHS = 10
BATCH_SIZE = 256

# ==========================================
# 5. TRAIN AND EVALUATE (WITH CUSTOM CV)
# ==========================================
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for model_name, build_fn in models.items():
    print(f"\n[{model_name}] Step 1: Running Custom 5-Fold CV on Train Set...")
    cv_time = time.time()

    fold_aurocs = []

    # Custom 5-Fold Loop
    for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_scaled, y_train)):
        X_fold_train, X_fold_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
        y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]

        # Build fresh model for each fold
        model = build_fn()
        model.fit(X_fold_train, y_fold_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=0)

        # Predict probabilities on validation fold
        y_val_pred_proba = model.predict(X_fold_val, verbose=0).ravel()
        fold_auc = roc_auc_score(y_fold_val, y_val_pred_proba)
        fold_aurocs.append(fold_auc)
        print(f"  -> Fold {fold+1} AUROC: {fold_auc:.4f}")

    mean_cv = np.mean(fold_aurocs)
    std_cv = np.std(fold_aurocs)
    print(f"[{model_name}] CV done in {(time.time() - cv_time)/60:.2f} minutes.")

    cv_performance['Model'].append(model_name)
    cv_performance['Parameters'].append(f"Epochs: {EPOCHS}, Batch: {BATCH_SIZE}")
    cv_performance['Fold 1 (AUROC)'].append(fold_aurocs[0])
    cv_performance['Fold 2 (AUROC)'].append(fold_aurocs[1])
    cv_performance['Fold 3 (AUROC)'].append(fold_aurocs[2])
    cv_performance['Fold 4 (AUROC)'].append(fold_aurocs[3])
    cv_performance['Fold 5 (AUROC)'].append(fold_aurocs[4])
    cv_performance['Mean CV AUROC'].append(mean_cv)
    cv_performance['Std CV AUROC'].append(std_cv)

    # --- Step 2: Train Final Model and Predict on Test Set ---
    print(f"[{model_name}] Step 2: Training final model for Test Set prediction...")
    final_model = build_fn()
    final_model.fit(X_train_scaled, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=0)

    # Keras returns shape (n, 1), .ravel() flattens it to (n,)
    y_pred_proba = final_model.predict(X_test_scaled, verbose=0).ravel()

    # Convert probabilities to binary predictions (threshold = 0.5)
    y_pred = (y_pred_proba > 0.5).astype(int)

    # --- Confusion Matrix Plotting & Saving ---
    cm = confusion_matrix(y_test, y_pred)

    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 14, "weight": "bold"})
    plt.title(f'{model_name} Confusion Matrix\n({sampling_name})', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')
    plt.tight_layout()

    safe_model_name = model_name.replace(' ', '_').replace('+', 'Plus')
    cm_filename = f"CM_{safe_model_name}_{sampling_name}.png"
    plt.savefig(os.path.join(cm_dir, cm_filename), dpi=300)
    plt.close()

    # --- Calculate Metrics ---
    report = classification_report(y_test, y_pred, output_dict=True)
    balanced_acc = balanced_accuracy_score(y_test, y_pred)
    AUROC = roc_auc_score(y_test, y_pred_proba)

    # Convert numpy array y_train back to Series just to count classes for logging
    train_counts = pd.Series(y_train).value_counts()

    precision_c0 = report['0']['precision'] if '0' in report else report['0.0']['precision']
    recall_c0 = report['0']['recall'] if '0' in report else report['0.0']['recall']
    f1_score_c0 = report['0']['f1-score'] if '0' in report else report['0.0']['f1-score']
    support_c0 = report['0']['support'] if '0' in report else report['0.0']['support']

    precision_c1 = report['1']['precision'] if '1' in report else report['1.0']['precision']
    recall_c1 = report['1']['recall'] if '1' in report else report['1.0']['recall']
    f1_score_c1 = report['1']['f1-score'] if '1' in report else report['1.0']['f1-score']
    support_c1 = report['1']['support'] if '1' in report else report['1.0']['support']

    train_class_0_size = train_counts.get(0, 0)
    train_class_1_size = train_counts.get(1, 0)

    # --- Print to Console ---
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
    print("Fill method:", data_name)
    print("Model:", model_name)
    print("Sampling:", sampling_name)
    print("Model Sampling:", model_name + ' ' + sampling_name)
    print("Class 0 #:", train_class_0_size)
    print("Class 1 #:", train_class_1_size)
    print("Precision c1:", precision_c1)
    print("Recall c1:", recall_c1)
    print("F1 Score c1:", f1_score_c1)
    print("Support c1:", support_c1)
    print("Precision c0:", precision_c0)
    print("Recall c0:", recall_c0)
    print("F1 Score c0:", f1_score_c0)
    print("Support c0:", support_c0)
    print("Overall Sensitivity:", sensitivity)
    print("Overall Specificity:", specificity)
    print("Balanced Accuracy:", balanced_acc)
    print("AUROC:", AUROC)
    print(f"Mean Train CV AUROC: {mean_cv:.4f} (+/- {std_cv:.4f})")
    print(f"Saved Confusion Matrix to Drive.")

    performance['Fill method'].append(data_name)
    performance['Model'].append(model_name)
    performance['Sampling'].append(sampling_name)
    performance['Model Sampling'].append(model_name + ' ' + sampling_name)
    performance['Class 0 #'].append(train_class_0_size)
    performance['Class 1 #'].append(train_class_1_size)
    performance['Precision c1'].append(precision_c1)
    performance['Recall c1'].append(recall_c1)
    performance['F1 Score c1'].append(f1_score_c1)
    performance['Support c1'].append(support_c1)
    performance['Precision c0'].append(precision_c0)
    performance['Recall c0'].append(recall_c0)
    performance['F1 Score c0'].append(f1_score_c0)
    performance['Support c0'].append(support_c0)
    performance['Overall Sensitivity'].append(sensitivity)
    performance['Overall Specificity'].append(specificity)
    performance['Balanced Accuracy'].append(balanced_acc)
    performance['AUROC'].append(AUROC)

    # ==========================================
    # 6. CONCATENATE & SAVE TO EXCEL (INSIDE LOOP)
    # ==========================================
    new_perf_df = pd.DataFrame(performance)
    new_cv_df = pd.DataFrame(cv_performance)

    if not base_perf_df.empty:
        save_perf_df = pd.concat([base_perf_df, new_perf_df], ignore_index=True)
        save_cv_df = pd.concat([base_cv_df, new_cv_df], ignore_index=True)
    else:
        save_perf_df = new_perf_df
        save_cv_df = new_cv_df

    with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
        save_perf_df.to_excel(writer, sheet_name='Test Metrics', index=False)
        save_cv_df.to_excel(writer, sheet_name='CV Scores', index=False)

    print(f"[{model_name}] Progress fully appended and saved to Excel.\n")

print("All Deep Learning models successfully completed and appended!")
```

```python
import pandas as pd
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import cross_validate, StratifiedKFold
from sklearn.metrics import classification_report, balanced_accuracy_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler

# ML Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# Deep Learning Models
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Dense, Dropout, MultiHeadAttention, Reshape, Flatten
import tensorflow.keras.backend as K

# ==========================================
# 1. SETUP PATHS AND LOAD DATA
# ==========================================
print("Loading GAN RUS datasets...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'

train_path = os.path.join(base_dir, 'GAN filling method/Resampled Datasets/3-BRFSS 2021 - Fill GAN RUS (processed).csv')
test_features_path = os.path.join(base_dir, 'GAN filling method/Test_Imputed_GAN_Processed.csv')
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')

cm_dir = os.path.join(base_dir, 'GAN filling method/Confusion Matrices/')
os.makedirs(cm_dir, exist_ok=True)
excel_output_path = os.path.join(base_dir, 'Performance_of_all_models.xlsx')

# Load Train & Test
df_train = pd.read_csv(train_path)
X_train = df_train.drop('Arthritis', axis=1).values
y_train = df_train['Arthritis'].values

df_test = pd.read_csv(test_features_path)
X_test = df_test.values
df_test_orig = pd.read_csv(original_test_path)
y_test = df_test_orig['Arthritis'].values

data_name = "GAN Imputation"
sampling_name = "RUS"

# ==========================================
# 2. FEATURE SCALING (FOR NEURAL NETS)
# ==========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ==========================================
# 3. INITIALIZE EXCEL APPENDING STRUCTURE
# ==========================================
if os.path.exists(excel_output_path):
    print("Found existing Performance Excel file. New results will be appended.")
    base_perf_df = pd.read_excel(excel_output_path, sheet_name='Test Metrics')
    base_cv_df = pd.read_excel(excel_output_path, sheet_name='CV Scores')
else:
    print("No existing Excel file found. A new one will be created.")
    base_perf_df = pd.DataFrame()
    base_cv_df = pd.DataFrame()

def get_empty_perf_dict():
    return {
        'Fill method': [], 'Model': [], 'Sampling': [], 'Model Sampling': [],
        'Class 0 #': [], 'Class 1 #': [],
        'Precision c1': [], 'Recall c1': [], 'F1 Score c1': [], 'Support c1': [],
        'Precision c0': [], 'Recall c0': [], 'F1 Score c0': [], 'Support c0': [],
        'Overall Sensitivity': [], 'Overall Specificity': [],
        'Balanced Accuracy': [], 'AUROC': []
    }

def get_empty_cv_dict():
    return {
        'Model': [], 'Parameters': [],
        'Fold 1 (AUROC)': [], 'Fold 2 (AUROC)': [], 'Fold 3 (AUROC)': [],
        'Fold 4 (AUROC)': [], 'Fold 5 (AUROC)': [],
        'Mean CV AUROC': [], 'Std CV AUROC': []
    }

# ==========================================
# 4. DEFINE TRADITIONAL & ENSEMBLE ML MODELS
# ==========================================
ml_models = {
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000, n_jobs=-1),
    "Random Forest": RandomForestClassifier(random_state=42, n_jobs=-1),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', n_jobs=-1),
    "LightGBM": LGBMClassifier(random_state=42, n_jobs=-1),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "AdaBoost": AdaBoostClassifier(random_state=42)
}

# Add Stacked Ensemble
stacking_estimators = [
    ('lgbm', LGBMClassifier(random_state=42, n_jobs=-1)),
    ('xgb', XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', n_jobs=-1)),
    ('gb', GradientBoostingClassifier(random_state=42))
]
ml_models["Stacked Ensemble"] = StackingClassifier(
    estimators=stacking_estimators,
    final_estimator=LogisticRegression(random_state=42, max_iter=1000),
    cv=5, n_jobs=-1
)

# ==========================================
# 5. DEFINE DEEP LEARNING ARCHITECTURES
# ==========================================
input_dim = X_train_scaled.shape[1]

def build_ann_1_layer():
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

def build_ann_2_layer():
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

def build_attention_ann():
    inputs = Input(shape=(input_dim,))
    x_reshaped = Reshape((input_dim, 1))(inputs)
    attention_out = MultiHeadAttention(num_heads=2, key_dim=2)(x_reshaped, x_reshaped)
    x = Flatten()(attention_out)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

dl_models = {
    "ANN 1 Layer": build_ann_1_layer,
    "ANN 2 Layer": build_ann_2_layer,
    "Attention + ANN 2 Layer": build_attention_ann
}

# ==========================================
# 6. MASTER EXECUTION FUNCTION
# ==========================================
def evaluate_and_save(model_name, mean_cv, std_cv, fold_scores, y_pred, y_pred_proba, params):
    global base_perf_df, base_cv_df

    performance = get_empty_perf_dict()
    cv_performance = get_empty_cv_dict()

    # Generate Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 14, "weight": "bold"})
    plt.title(f'{model_name} Confusion Matrix\n({sampling_name})', fontsize=14, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')
    plt.tight_layout()

    safe_model_name = model_name.replace(' ', '_').replace('+', 'Plus')
    cm_filename = f"CM_{safe_model_name}_{sampling_name}.png"
    plt.savefig(os.path.join(cm_dir, cm_filename), dpi=300)
    plt.close()

    # Calculate Metrics
    report = classification_report(y_test, y_pred, output_dict=True)
    balanced_acc = balanced_accuracy_score(y_test, y_pred)
    AUROC = roc_auc_score(y_test, y_pred_proba)

    train_counts = pd.Series(y_train).value_counts()

    precision_c0 = report['0']['precision'] if '0' in report else report['0.0']['precision']
    recall_c0 = report['0']['recall'] if '0' in report else report['0.0']['recall']
    f1_score_c0 = report['0']['f1-score'] if '0' in report else report['0.0']['f1-score']
    support_c0 = report['0']['support'] if '0' in report else report['0.0']['support']

    precision_c1 = report['1']['precision'] if '1' in report else report['1.0']['precision']
    recall_c1 = report['1']['recall'] if '1' in report else report['1.0']['recall']
    f1_score_c1 = report['1']['f1-score'] if '1' in report else report['1.0']['f1-score']
    support_c1 = report['1']['support'] if '1' in report else report['1.0']['support']

    train_class_0_size = train_counts.get(0, 0)
    train_class_1_size = train_counts.get(1, 0)

    # Print Console Output
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
    print("Fill method:", data_name)
    print("Model:", model_name)
    print("Sampling:", sampling_name)
    print("Model Sampling:", model_name + ' ' + sampling_name)
    print("Class 0 #:", train_class_0_size)
    print("Class 1 #:", train_class_1_size)
    print("Precision c1:", precision_c1)
    print("Recall c1:", recall_c1)
    print("F1 Score c1:", f1_score_c1)
    print("Support c1:", support_c1)
    print("Precision c0:", precision_c0)
    print("Recall c0:", recall_c0)
    print("F1 Score c0:", f1_score_c0)
    print("Support c0:", support_c0)
    print("Overall Sensitivity:", sensitivity)
    print("Overall Specificity:", specificity)
    print("Balanced Accuracy:", balanced_acc)
    print("AUROC:", AUROC)
    print(f"Mean Train CV AUROC: {mean_cv:.4f} (+/- {std_cv:.4f})")
    print(f"Saved Confusion Matrix to Drive.")

    # Append Data
    performance['Fill method'].append(data_name)
    performance['Model'].append(model_name)
    performance['Sampling'].append(sampling_name)
    performance['Model Sampling'].append(model_name + ' ' + sampling_name)
    performance['Class 0 #'].append(train_class_0_size)
    performance['Class 1 #'].append(train_class_1_size)
    performance['Precision c1'].append(precision_c1)
    performance['Recall c1'].append(recall_c1)
    performance['F1 Score c1'].append(f1_score_c1)
    performance['Support c1'].append(support_c1)
    performance['Precision c0'].append(precision_c0)
    performance['Recall c0'].append(recall_c0)
    performance['F1 Score c0'].append(f1_score_c0)
    performance['Support c0'].append(support_c0)
    performance['Overall Sensitivity'].append(sensitivity)
    performance['Overall Specificity'].append(specificity)
    performance['Balanced Accuracy'].append(balanced_acc)
    performance['AUROC'].append(AUROC)

    cv_performance['Model'].append(model_name)
    cv_performance['Parameters'].append(params)
    cv_performance['Fold 1 (AUROC)'].append(fold_scores[0])
    cv_performance['Fold 2 (AUROC)'].append(fold_scores[1])
    cv_performance['Fold 3 (AUROC)'].append(fold_scores[2])
    cv_performance['Fold 4 (AUROC)'].append(fold_scores[3])
    cv_performance['Fold 5 (AUROC)'].append(fold_scores[4])
    cv_performance['Mean CV AUROC'].append(mean_cv)
    cv_performance['Std CV AUROC'].append(std_cv)

    # Save to Excel
    new_perf_df = pd.DataFrame(performance)
    new_cv_df = pd.DataFrame(cv_performance)

    if not base_perf_df.empty:
        base_perf_df = pd.concat([base_perf_df, new_perf_df], ignore_index=True)
        base_cv_df = pd.concat([base_cv_df, new_cv_df], ignore_index=True)
    else:
        base_perf_df = new_perf_df
        base_cv_df = new_cv_df

    with pd.ExcelWriter(excel_output_path, engine='openpyxl') as writer:
        base_perf_df.to_excel(writer, sheet_name='Test Metrics', index=False)
        base_cv_df.to_excel(writer, sheet_name='CV Scores', index=False)

    print(f"[{model_name}] Progress fully appended and saved to Excel.\n")

# ==========================================
# 7. RUN TRADITIONAL & ENSEMBLE MODELS
# ==========================================
print("\n=== STARTING ML MODELS ===")
for model_name, model in ml_models.items():
    print(f"\n[{model_name}] Step 1: Running 5-Fold CV...")
    cv_time = time.time()

    cv_res = cross_validate(model, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1)
    folds = cv_res['test_score']
    mean_cv = folds.mean()
    std_cv = folds.std()
    print(f"[{model_name}] CV done in {(time.time() - cv_time)/60:.2f} minutes.")

    print(f"[{model_name}] Step 2: Training final model on Test Set...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    params = "Base: LGBM, XGB, GB | Meta: LR" if "Stacked" in model_name else "Default"
    evaluate_and_save(model_name, mean_cv, std_cv, folds, y_pred, y_pred_proba, params)

# ==========================================
# 8. RUN DEEP LEARNING MODELS
# ==========================================
print("\n=== STARTING DEEP LEARNING MODELS ===")
EPOCHS = 10
BATCH_SIZE = 256
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for model_name, build_fn in dl_models.items():
    print(f"\n[{model_name}] Step 1: Running Custom 5-Fold CV...")
    cv_time = time.time()
    fold_aurocs = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X_train_scaled, y_train)):
        K.clear_session() # Clear GPU memory
        X_fold_train, X_fold_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
        y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]

        model = build_fn()
        model.fit(X_fold_train, y_fold_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=0)

        y_val_pred_proba = model.predict(X_fold_val, verbose=0).ravel()
        fold_auc = roc_auc_score(y_fold_val, y_val_pred_proba)
        fold_aurocs.append(fold_auc)
        print(f"  -> Fold {fold+1} AUROC: {fold_auc:.4f}")

    mean_cv = np.mean(fold_aurocs)
    std_cv = np.std(fold_aurocs)
    print(f"[{model_name}] CV done in {(time.time() - cv_time)/60:.2f} minutes.")

    print(f"[{model_name}] Step 2: Training final model on Test Set...")
    K.clear_session()
    final_model = build_fn()
    final_model.fit(X_train_scaled, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=0)

    y_pred_proba = final_model.predict(X_test_scaled, verbose=0).ravel()
    y_pred = (y_pred_proba > 0.5).astype(int)

    params = f"Epochs: {EPOCHS}, Batch: {BATCH_SIZE}"
    evaluate_and_save(model_name, mean_cv, std_cv, fold_aurocs, y_pred, y_pred_proba, params)

print("\n🎉 ENTIRE PIPELINE SUCCESSFULLY COMPLETED FOR ALL 11 MODELS!")
```

```python
import pandas as pd
import numpy as np
import os
import joblib
import shap
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, confusion_matrix, balanced_accuracy_score
import warnings
warnings.filterwarnings('ignore')

# ==========================================
# 1. SETUP PATHS & LOAD DATA
# ==========================================
print("Initializing Unified XGBoost Pipeline (Train + Metrics + Regrouped SHAP)...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'
method_dir = os.path.join(base_dir, 'GAN filling method')
models_dir = os.path.join(method_dir, 'Saved_Models')
shap_out_dir = os.path.join(base_dir, 'SHAP_Visualizations')

os.makedirs(models_dir, exist_ok=True)
os.makedirs(shap_out_dir, exist_ok=True)

# ---> 1. Load Training Data (GAN + ROS) <---
train_filename = '3-BRFSS 2021 - Fill GAN ROS (processed).csv'
train_path = os.path.join(method_dir, 'Resampled Datasets', train_filename)
df_train = pd.read_csv(train_path)
X_train = df_train.drop('Arthritis', axis=1).values
y_train = df_train['Arthritis'].values

# ---> 2. Load Processed Test Features (for model prediction and SHAP) <---
test_features_path = os.path.join(method_dir, 'Test_Imputed_GAN_Processed.csv')
X_test_df = pd.read_csv(test_features_path)
if 'Arthritis' in X_test_df.columns:
    X_test_df = X_test_df.drop('Arthritis', axis=1)
X_test_vals = X_test_df.values

# ---> 3. Load Original Test Data (for clean demographic slicing and true labels) <---
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')
df_test_orig = pd.read_csv(original_test_path)
y_test = df_test_orig['Arthritis'].values

# ==========================================
# 2. TRAIN & SAVE CHAMPION MODEL
# ==========================================
print("\nTraining XGBoost Champion Model...")
model = XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss',
    n_jobs=-1
)

model.fit(X_train, y_train)

# Save the model
model_save_path = os.path.join(models_dir, 'XGBoost_ROS.pkl')
joblib.dump(model, model_save_path)
print(f"🎉 Champion Model saved to: {model_save_path}")

# ==========================================
# 3. DEFINING COHORTS (Safe Slicing)
# ==========================================
GENDER_COL = 'Gender'
AGE_COL = 'Age'

# Grab row indices based on original data to avoid dummy variable KeyError
cohort_indices = {
    "Overall": df_test_orig.index.tolist(),
    "Male": df_test_orig.index[df_test_orig[GENDER_COL] == 0].tolist(),
    "Female": df_test_orig.index[df_test_orig[GENDER_COL] == 1].tolist(),
    "Younger (Age < 8)": df_test_orig.index[df_test_orig[AGE_COL] < 8].tolist(),
    "Older (Age >= 8)": df_test_orig.index[df_test_orig[AGE_COL] >= 8].tolist()
}

# ==========================================
# 4. SUBGROUP PERFORMANCE METRICS
# ==========================================
print("\n" + "="*50)
print(" CLINICAL PERFORMANCE METRICS")
print("="*50)

def evaluate_subgroup(group_name, indices):
    if len(indices) == 0: return

    X_sub = X_test_vals[indices]
    y_sub = y_test[indices]

    y_pred = model.predict(X_sub)
    y_pred_proba = model.predict_proba(X_sub)[:, 1]

    AUROC = roc_auc_score(y_sub, y_pred_proba)
    bal_acc = balanced_accuracy_score(y_sub, y_pred)

    cm = confusion_matrix(y_sub, y_pred)
    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

    print(f"--- Performance for {group_name} (n={len(y_sub)}) ---")
    print(f"AUROC: {AUROC:.4f} | Bal Acc: {bal_acc:.4f} | Sens: {sensitivity:.4f} | Spec: {specificity:.4f}\n")

for name, idx in cohort_indices.items():
    evaluate_subgroup(name, idx)

# ==========================================
# 5. FEATURE REGROUPING LOGIC
# ==========================================
def regroup_shap_and_features(shap_matrix, original_df):
    """Sums SHAP values for one-hot encoded columns back into their original feature, ignoring _nan columns."""
    grouped_cols = {}

    for idx, col in enumerate(original_df.columns):
        # Explicitly skip any column that represents a NaN dummy variable
        if col.endswith('_nan'):
            continue

        parts = col.rsplit('_', 1)
        if len(parts) == 2 and parts[1].replace('.', '', 1).isdigit():
            base_name = parts[0]
        else:
            base_name = col

        if base_name not in grouped_cols:
            grouped_cols[base_name] = []
        grouped_cols[base_name].append(idx)

    new_shap = np.zeros((shap_matrix.shape[0], len(grouped_cols)))
    new_features = pd.DataFrame(index=original_df.index)

    for i, (base_name, indices) in enumerate(grouped_cols.items()):
        new_shap[:, i] = np.sum(shap_matrix[:, indices], axis=1)

        if len(indices) > 1:
            col_names = [original_df.columns[idx] for idx in indices]
            series = pd.Series(0.0, index=original_df.index)
            for col in col_names:
                cat_val = float(col.rsplit('_', 1)[1])
                series += original_df[col] * cat_val
            new_features[base_name] = series
        else:
            new_features[base_name] = original_df.iloc[:, indices[0]]

    return new_shap, new_features

# ==========================================
# 6. FAST SHAP GENERATION & PLOTTING
# ==========================================
print("="*50)
print(" SHAP GENERATION (TREE EXPLAINER)")
print("="*50)

explainer = shap.TreeExplainer(model)

shap_cohorts = ["Overall", "Male", "Female"]

for cohort_name in shap_cohorts:
    indices = cohort_indices[cohort_name]
    if len(indices) == 0: continue

    print(f"\n-> Processing SHAP for {cohort_name} Cohort (n={len(indices)})...")

    X_cohort_df = X_test_df.iloc[indices].copy()
    shap_values_obj = explainer.shap_values(X_cohort_df)

    if isinstance(shap_values_obj, list):
        raw_shap_values = shap_values_obj[1]
    else:
        raw_shap_values = shap_values_obj

    print("   Regrouping one-hot encoded features and removing NaNs...")
    regrouped_shap, regrouped_features = regroup_shap_and_features(raw_shap_values, X_cohort_df)

    safe_filename = f"BRFSS_Regrouped_{cohort_name}_XGBoost_ROS"

    num_features = len(regrouped_features.columns)
    plot_height = max(10, num_features * 0.35)

    # --- 1. BAR PLOT WITH ANNOTATIONS ---
    print("   Saving Bar Plot...")
    plt.figure(figsize=(10, plot_height))
    shap.summary_plot(regrouped_shap, regrouped_features, plot_type="bar", show=False, color="#ff0051", max_display=num_features)

    ax = plt.gca()
    # Annotate bars
    for p in ax.patches:
        width = p.get_width()
        if width > 0:
            ax.text(width + (ax.get_xlim()[1] * 0.01), p.get_y() + p.get_height() / 2,
                    f'+{width:.2f}', ha='left', va='center', color="#ff0051", fontweight='bold', fontsize=11)

    # Force feature names (y-tick labels) to be bold
    for label in ax.get_yticklabels():
        label.set_fontweight('bold')
        label.set_fontsize(11)

    plt.title(f"BRFSS Global Feature Importance ({cohort_name})\nXGBoost (GAN ROS)", fontweight='bold', fontsize=14)
    plt.xlim(0, ax.get_xlim()[1] * 1.1)
    plt.tight_layout()
    plt.savefig(os.path.join(shap_out_dir, f"{safe_filename}_Bar.png"), dpi=300, bbox_inches='tight')
    plt.close()

    # --- 2. BEESWARM PLOT ---
    print("   Saving Beeswarm Plot...")
    plt.figure(figsize=(10, plot_height))
    shap.summary_plot(regrouped_shap, regrouped_features, show=False, max_display=num_features)

    ax = plt.gca()
    # Force feature names (y-tick labels) to be bold on beeswarm as well
    for label in ax.get_yticklabels():
        label.set_fontweight('bold')
        label.set_fontsize(11)

    plt.title(f"BRFSS SHAP Summary ({cohort_name})\nXGBoost (GAN ROS)", fontweight='bold', fontsize=14)
    plt.tight_layout()
    plt.savefig(os.path.join(shap_out_dir, f"{safe_filename}_Beeswarm.png"), dpi=300, bbox_inches='tight')
    plt.close()

print("\n🎉 MASTER PIPELINE COMPLETE! Metrics printed and all final, clean images saved to Drive.")
```

```python
import pandas as pd
import numpy as np
import os
import joblib
import shap
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# ==========================================
# 1. SETUP & LOAD DATA
# ==========================================
print("Initializing MAX Absolute SHAP & Subgroup Analysis for BRFSS...")
base_dir = '/content/drive/MyDrive/Arthritis version 2/BRFSS 2021/'
method_dir = os.path.join(base_dir, 'GAN filling method')
shap_out_dir = os.path.join(base_dir, 'SHAP_Visualizations')
os.makedirs(shap_out_dir, exist_ok=True)

# Load Model
model_path = os.path.join(method_dir, 'Saved_Models', 'XGBoost_ROS.pkl')
model = joblib.load(model_path)

# Load Processed Features
test_features_path = os.path.join(method_dir, 'Test_Imputed_GAN_Processed.csv')
X_test_df = pd.read_csv(test_features_path)
if 'Arthritis' in X_test_df.columns:
    X_test_df = X_test_df.drop('Arthritis', axis=1)

# Load Original Test Data (for clean demographic slicing)
original_test_path = os.path.join(base_dir, 'BRFSS_2021_Test.csv')
df_test_orig = pd.read_csv(original_test_path)

# ==========================================
# 2. DEFINING COHORTS (Safe Slicing)
# ==========================================
GENDER_COL = 'Gender'

cohort_indices = {
    "Overall": df_test_orig.index.tolist(),
    "Male": df_test_orig.index[df_test_orig[GENDER_COL] == 0].tolist(),
    "Female": df_test_orig.index[df_test_orig[GENDER_COL] == 1].tolist()
}

# ==========================================
# 3. FEATURE REGROUPING LOGIC
# ==========================================
def regroup_shap_and_features(shap_matrix, original_df):
    """Sums SHAP values for one-hot encoded columns back into their original feature, ignoring _nan columns."""
    grouped_cols = {}
    for idx, col in enumerate(original_df.columns):
        if col.endswith('_nan'): continue
        parts = col.rsplit('_', 1)
        if len(parts) == 2 and parts[1].replace('.', '', 1).isdigit():
            base_name = parts[0]
        else:
            base_name = col

        if base_name not in grouped_cols:
            grouped_cols[base_name] = []
        grouped_cols[base_name].append(idx)

    new_shap = np.zeros((shap_matrix.shape[0], len(grouped_cols)))
    new_features = pd.DataFrame(index=original_df.index)

    for i, (base_name, indices) in enumerate(grouped_cols.items()):
        new_shap[:, i] = np.sum(shap_matrix[:, indices], axis=1)
        if len(indices) > 1:
            col_names = [original_df.columns[idx] for idx in indices]
            series = pd.Series(0.0, index=original_df.index)
            for col in col_names:
                cat_val = float(col.rsplit('_', 1)[1])
                series += original_df[col] * cat_val
            new_features[base_name] = series
        else:
            new_features[base_name] = original_df.iloc[:, indices[0]]

    return new_shap, new_features

# ==========================================
# 4. MASTER SHAP GENERATION LOOP
# ==========================================
print(" -> Setting up TreeExplainer...")
explainer = shap.TreeExplainer(model)

shap_cohorts = ["Overall", "Male", "Female"]

for cohort_name in shap_cohorts:
    indices = cohort_indices[cohort_name]
    if len(indices) == 0: continue

    print(f"\n{'='*50}")
    print(f" PROCESSING MAX SHAP: {cohort_name} COHORT (n={len(indices)})")
    print(f"{'='*50}")

    # Slice the specific cohort features
    X_cohort_df = X_test_df.iloc[indices].copy()

    # Calculate SHAP values
    shap_values_obj = explainer.shap_values(X_cohort_df)
    if isinstance(shap_values_obj, list):
        raw_shap_values = shap_values_obj[1]
    else:
        raw_shap_values = shap_values_obj

    print(" -> Regrouping features and filtering NaNs...")
    regrouped_shap, regrouped_features = regroup_shap_and_features(raw_shap_values, X_cohort_df)

    # Calculate Max SHAP
    print(" -> Extracting Maximum Absolute SHAP per feature...")
    max_shap_vals = np.max(np.abs(regrouped_shap), axis=0)
    feature_names = regrouped_features.columns

    # ----------------------------------------------------
    # ALIGNMENT MATH: Sort features by MAX SHAP
    # ----------------------------------------------------
    sort_idx = np.argsort(max_shap_vals)

    # 1. Sort ascending for the Bar chart (plt.barh draws bottom-to-top)
    sorted_max_shap = max_shap_vals[sort_idx]
    sorted_features = feature_names[sort_idx]

    # 2. Sort descending for the Beeswarm chart (shap.summary_plot draws top-to-bottom)
    sort_idx_reversed = sort_idx[::-1]
    regrouped_shap_sorted = regrouped_shap[:, sort_idx_reversed]
    regrouped_features_sorted = regrouped_features.iloc[:, sort_idx_reversed]

    safe_filename = f"BRFSS_{cohort_name}_XGBoost_ROS_MAX"
    num_features = len(sorted_features)

    # Strictly define height so both plots are identical
    plot_height = max(10, num_features * 0.35)

    # ----------------------------------------------------
    # 1. PLOT: MAX SHAP BAR CHART (Q1 Polished)
    # ----------------------------------------------------
    print(" -> Generating Annotated Max SHAP Bar Plot...")
    plt.figure(figsize=(10, plot_height))

    bars = plt.barh(np.arange(num_features), sorted_max_shap, color="#ff0051", edgecolor='none')

    # Bold feature names on y-axis
    plt.yticks(np.arange(num_features), sorted_features, fontweight='bold', fontsize=12)
    plt.xticks(fontsize=11)

    # Add numeric annotations beside the bars
    for i, bar in enumerate(bars):
        width = bar.get_width()
        plt.text(width + (max(sorted_max_shap) * 0.01), bar.get_y() + bar.get_height()/2,
                 f'+{width:.2f}', va='center', color="#ff0051", fontweight='bold', fontsize=11)

    # Clean axes and bold x-label
    plt.xlabel("max(|SHAP value|)", fontweight='bold', fontsize=12)
    plt.xlim(0, max(sorted_max_shap) * 1.15)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    plt.tight_layout()
    plt.savefig(os.path.join(shap_out_dir, f"{safe_filename}_SHAP_Bar.png"), dpi=300, bbox_inches='tight')
    plt.close()

    # ----------------------------------------------------
    # 2. PLOT: MAX-SORTED BEESWARM (Q1 Polished)
    # ----------------------------------------------------
    print(" -> Generating Max-Sorted Beeswarm Plot...")
    plt.figure(figsize=(10, plot_height))

    # sort=False forces SHAP to use our custom Max-sorting logic
    shap.summary_plot(regrouped_shap_sorted, regrouped_features_sorted, show=False, sort=False, max_display=num_features)

    ax = plt.gca()

    # Bold feature names on y-axis for beeswarm
    for label in ax.get_yticklabels():
        label.set_fontweight('bold')
        label.set_fontsize(12)

    for label in ax.get_xticklabels():
        label.set_fontsize(11)

    # Bold x-label to match the bar chart styling
    ax.set_xlabel("SHAP value (impact on model output)", fontweight='bold', fontsize=12)

    plt.tight_layout()
    plt.savefig(os.path.join(shap_out_dir, f"{safe_filename}_SHAP_Beeswarm.png"), dpi=300, bbox_inches='tight')
    plt.close()

print("\n🎉 ALL Q1-STANDARD MAX SHAP PLOTS (BAR & BEESWARM) SUCCESSFULLY GENERATED!")
```
